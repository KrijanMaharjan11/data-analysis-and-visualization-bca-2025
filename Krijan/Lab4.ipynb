Exploring 1D Data Visualizations

```python
# Import required libraries at the top
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load your dataset
df = pd.read_csv("train_and_test2.csv")

# Display first few rows
df.head()

```

1. Histograms (for Numerical Data)

Standard Histogram â€“ Age

```python
plt.figure(figsize=(8, 5))
plt.hist(df['Age'].dropna(), bins=30, color='skyblue', edgecolor='black')
plt.title('Histogram of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()

```

1.2 Normalized Histogram â€“ Fare

```python
plt.figure(figsize=(8, 5))
plt.hist(df['Fare'], bins=30, density=True, color='orange', edgecolor='black')
plt.title('Normalized Histogram of Fare')
plt.xlabel('Fare')
plt.ylabel('Density')
plt.grid(True)
plt.show()
```

1.3 Cumulative Histogram â€“ Age

```python
plt.figure(figsize=(8, 5))
plt.hist(df['Age'].dropna(), bins=30, cumulative=True, color='lightgreen', edgecolor='black')
plt.title('Cumulative Histogram of Age')
plt.xlabel('Age')
plt.ylabel('Cumulative Count')
plt.grid(True)
plt.show()
```

1.4 KDE / Density Plot (Seaborn)

```python
plt.figure(figsize=(8, 5))
sns.kdeplot(df['Fare'], fill=True, color='purple')
plt.title('Density Plot of Fare')
plt.xlabel('Fare')
plt.ylabel('Density')
plt.grid(True)
plt.show()

```

**2. Bar Plots (for Categorical Data)**

2.1 Vertical Bar Chart â€“ Sex

```python
df['Sex'].value_counts().plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Count of Passengers by Sex')
plt.xlabel('Sex')
plt.ylabel('Count')
plt.grid(True, axis='y')
plt.show()

```

   2.2 Stacked Bar Chart â€“ Sex vs. Survival

```python
pd.crosstab(df['Sex'], df['2urvived']).plot(kind='bar', stacked=True, color=['red', 'green'])

plt.title('Stacked Bar Chart of Sex vs 2urvived')
plt.xlabel('Sex')
plt.ylabel('Count')
plt.legend(title='2urvived', labels=['Did Not Survive', 'Survived'])  # Adjust labels if needed
plt.show()
```

2.3 Grouped Bar Chart â€“ Pclass vs. Survived

```python
pd.crosstab(df['Pclass'], df['2urvived']).plot(kind='bar', stacked=False, color=['red', 'green'])

plt.title('Grouped Bar Chart of Pclass vs 2urvived')
plt.xlabel('Passenger Class')
plt.ylabel('Count')
plt.legend(title='2urvived', labels=['Did Not Survive', 'Survived'])  # Adjust labels as needed
plt.show()

```

2.4 Lollipop Chart â€“ Embarked

```python
counts = df['Embarked'].value_counts()
plt.figure(figsize=(6,4))
plt.stem(counts.index, counts.values, basefmt=" ")
plt.title("Lollipop Chart of Embarked Locations")
plt.xlabel("Port")
plt.ylabel("Count")
plt.show()
```

3. Pie Charts (for Category Proportion)

3.1 Standard Pie Chart â€“ Sex

```python
df['Sex'].value_counts().plot.pie(autopct='%1.1f%%')
plt.title("Gender Distribution")
plt.ylabel('')
plt.show()
```

3.2 Donut Chart â€“ Embarked

```python
embarked_counts = df['Embarked'].value_counts()
plt.pie(embarked_counts, labels=embarked_counts.index, autopct='%1.1f%%', wedgeprops=dict(width=0.4))
plt.title("Donut Chart of Embarked")
plt.show()
```

3.3 Exploded Pie Chart â€“ Pclass

```python
pclass_counts = df['Pclass'].value_counts()
explode = (0.1, 0, 0)
plt.pie(pclass_counts, labels=pclass_counts.index, explode=explode, autopct='%1.1f%%')
plt.title("Exploded Pie Chart of Passenger Classes")
plt.show()
```

3.4 Nested Pie (Sunburst style) â€“ Sex and Survived

```python
import plotly.express as px

fig = px.sunburst(df, path=['Sex', '2urvived'], values=None)
fig.update_layout(title="Nested Pie (Sunburst) of Sex and 2urvived")
fig.show()
```

4. Box Plots (for Statistical Summary)

4.1 Standard Box Plot â€“ Age

```python
plt.figure(figsize=(6,5))
sns.boxplot(y='Age', data=df)
plt.title("Box Plot of Age")
plt.show()
```

4.2 Notched Box Plot â€“ Fare

```python
plt.figure(figsize=(6,5))
sns.boxplot(y='Fare', data=df, notch=True)
plt.title("Notched Box Plot of Fare")
plt.show()
```

4.3 Violin Plot â€“ Age

```python
plt.figure(figsize=(6,5))
sns.violinplot(y='Age', data=df)
plt.title("Violin Plot of Age")
plt.show()
```

4.4 Side-by-side Box Plot â€“ Age by Sex

```python
plt.figure(figsize=(6,5))
sns.boxplot(x='Sex', y='Age', data=df)
plt.title("Box Plot of Age by Sex")
plt.show()
```
