1. Sankey Diagram Example

```python
import matplotlib.pyplot as plt
from matplotlib.sankey import Sankey

# Example simplified flow (you might want to aggregate flows first)
flows = [500, 300, -500, -300]  # positive = input, negative = output
labels = ['Region A -> X', 'Region A -> Y', 'Region A out', 'Region A out']

fig, ax = plt.subplots(figsize=(8,6))
sankey = Sankey(ax=ax, scale=0.01)

# add flows: positive = inputs, negative = outputs
sankey.add(flows=flows, labels=labels, orientations=[1, 1, -1, -1])
sankey.finish()

plt.title('Basic Sankey Diagram with Matplotlib')
plt.show()
```

2.  Alluvial Plot

```python
import matplotlib.pyplot as plt
from matplotlib.sankey import Sankey

fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(1, 1, 1, xticks=[], yticks=[], title="Simple Alluvial (Sankey) Plot")

# Example flows: positive numbers are inputs, negatives outputs
# Adjust these numbers and labels as needed for your data
flows = [20, 30, 10, -15, -10, -35]  
labels = ['High School', 'Bachelor', 'Master', 'Clerical', 'Technical', 'Management']
orientations = [1, 1, 1, -1, -1, -1]  # 1 = flows up, -1 = flows down

sankey = Sankey(ax=ax, unit=None)
sankey.add(flows=flows, labels=labels, orientations=orientations)
sankey.finish()

plt.tight_layout()
plt.show()
```

3. Chord Diagram

```python
import pandas as pd
import holoviews as hv
from holoviews import opts
hv.extension('bokeh')

# Sample symmetric flow matrix
data = {
    'A': [0, 10, 5, 2],
    'B': [10, 0, 3, 7],
    'C': [5, 3, 0, 8],
    'D': [2, 7, 8, 0]
}
df = pd.DataFrame(data, index=['A', 'B', 'C', 'D'])

# Convert matrix to edge list
edges = []
for i in df.index:
    for j in df.columns:
        if df.loc[i, j] > 0:
            edges.append((i, j, df.loc[i, j]))

# Create Chord Diagram
chord = hv.Chord(edges)
chord.opts(
    opts.Chord(
        labels='index',
        cmap='Category20',
        edge_color='source',
        edge_cmap='Category20',
        node_color='index',
        edge_line_width=hv.dim('value') * 0.1,
        width=600, height=600, title="Chord Diagram"
    )
)
```

4. Directed Graph

```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.DiGraph()

edges = [("Region A", "Region X", 500), ("Region A", "Region Y", 300),
         ("Region B", "Region X", 200), ("Region B", "Region Z", 400)]

G.add_weighted_edges_from(edges)

pos = nx.spring_layout(G, seed=42)  # For consistent layout

edge_weights = [G[u][v]['weight'] / 100 for u,v in G.edges()]  # scaled for width

nx.draw(G, pos, with_labels=True, node_color='skyblue', node_size=2000,
        arrowsize=20, width=edge_weights, edge_color='gray')

edge_labels = {(u, v): d['weight'] for u, v, d in G.edges(data=True)}
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels)
plt.title("Directed Graph of Migration Flows")
plt.show()
```

5. Steam Graph

```python
import matplotlib.pyplot as plt
import numpy as np

years = [2020, 2021, 2022]

region_a_to_x = [500, 550, 600]
region_a_to_y = [300, 280, 320]
region_b_to_x = [200, 220, 250]
region_b_to_z = [400, 450, 470]

plt.stackplot(years, region_a_to_x, region_a_to_y, region_b_to_x, region_b_to_z,
              labels=["A→X", "A→Y", "B→X", "B→Z"], alpha=0.8)

plt.legend(loc='upper left')
plt.title("Stream Graph of Migration Flows Over Time")
plt.xlabel("Year")
plt.ylabel("Number of Migrants")
plt.show()
```
